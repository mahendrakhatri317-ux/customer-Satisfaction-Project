```python
sns.countplot(x='satisfaction', data=df)
plt.title('Customer Satisfaction Distribution')
plt.show()
```

```python
sns.countplot(x='Customer Type', data=df)
plt.title('Customer Type Distribution')
plt.show()
```

```python
sns.countplot(x='Gender' , hue='satisfaction', data=df)
plt.title('Customer Satisfaction by Gender')
plt.show()
```

```python
sns.histplot([df['Age']], bins=30, kde=True)
plt.title('Age Distribution')
plt.show()
```

```python
sns.scatterplot(x='Age', y='Cleanliness', data=df)
plt.show()
```

```python
sns.countplot(x='Class', hue='satisfaction', data=df)
plt.title('Class Distribution by Satisfaction')
plt.show()
```

```python
sns.countplot(x='Type of Travel', hue='satisfaction', data=df)
plt.title('Type of Travel Distribution by Satisfaction')
plt.show()
```

```python
sns.boxplot(x="satisfaction", y="Flight Distance", data=df)
plt.title('Flight Distance by Satisfaction')
plt.show()
```

```python
numeric_df = df.select_dtypes(include=['int64', 'float64'])
plt.figure(figsize=(10, 8))
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm', )
plt.title('Correlation Heatmap')
plt.show()
```
